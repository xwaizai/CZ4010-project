```python
def convertHexToBinary(h):
    num = int(h,16)
    binary_string = format(num, 'b')
    #making sure that the binary string is in length of mod 8
    while len(binary_string)%8 != 0:
        binary_string = "0" + binary_string
    return binary_string
```

```python
def convertBinaryToHex(n):
   
    # convert binary to int
    num = int(n, 2)
     
    # convert int to hexadecimal
    hex_num = format(num, 'x')
    return(hex_num)
```

```python
def initS():
    n=8
    S = [i for i in range(0, 2**n)]
    return S
```

```python
def KSA(keyBinary):
    n = 8

    # The initial state vector array
    S = initS()
    print("S : ", S)
 
    key_list = [keyBinary[i:i + n] for i in range(0, len(keyBinary), n)]
 
    # Convert to key_stream to decimal
    for i in range(len(key_list)):
        key_list[i] = int(key_list[i], 2)
 
    # Making key_stream equal
    # to length of state vector
    diff = int(len(S)-len(key_list))
 
    if diff != 0:
        for i in range(0, diff):
            key_list.append(key_list[i])
 
    print("Key list : ", key_list)
    print(" ")

    j = 0
    N = len(S)
        
    # Iterate over the range [0, N]
    for i in range(0, N):
        
        # Find the key
        j = (j + S[i]+key_list[i]) % N
            
        # Update S[i] and S[j]
        S[i], S[j] = S[j], S[i]
        # print(i, " ", end ="")
            
        # # Print S
        # print(S)

    initial_permutation_array = S
        
    print(" ")
    print("The initial permutation array is : ",
            initial_permutation_array)
    return S
    
```

```python
def PGRA(S,pt):
 
    N = len(S)
    i = j = 0
    key_stream = []

    # Iterate over [0, length of pt]
    for k in range(0, len(pt)):
        i = (i + 1) % N
        j = (j + S[i]) % N
            
        # Update S[i] and S[j]
        S[i], S[j] = S[j], S[i]
        # print(k, " ", end ="")
        # print(S)
        t = (S[i]+S[j]) % N
        key_stream.append(S[t])

    # Print the key stream
    print("Key stream : ", key_stream)
    print(" ")
    return key_stream
```

```python
# Performing XOR between generated
# key stream and plain text
def XOR(key_stream,pt):
    cipher_text = []
    for i in range(len(pt)):
        c = key_stream[i] ^ pt[i]
        cipher_text.append(c)
    return cipher_text
```

```python
def encryption(key, plain_text):
 
    # Given text (in hex) and key
    ptBinary = convertHexToBinary(plain_text)
    keyBinary = convertHexToBinary(key)
 
    # n is the no: of bits to
    # be considered at a time
    n = 8
 
    print("Plain text : ", plain_text)
    print("Key : ", key)
    print("Plain text binary: ", ptBinary)
    print("Key binary: ", keyBinary)
    print("n : ", n)
 
    print(" ")

    # Convert to plain_text to decimal
    pt = [ptBinary[i:i + n] for i in range(0, len(ptBinary), n)]
 
    for i in range(len(pt)):
        pt[i] = int(pt[i], 2)
 
    print("Plain text ( in array form ): ", pt)
 
 
 
    print("KSA iterations : ")
    print(" ")
    S = KSA(keyBinary)
    print(" ")
 
    # Perform PGRA algorithm
    
 
    print("PGRA iterations : ")
    print(" ")
    key_stream = PGRA(S, pt)
 
    
 
    cipher_text = XOR(key_stream,pt)
 
    # Convert the encrypted text to
    # bits form
    encrypted_to_bits = ""
    for i in cipher_text:
        encrypted_to_bits += '0'*(n-len(bin(i)[2:]))+bin(i)[2:]
 
    print(" ")
    print(key_stream)
    print("Cipher text : ", encrypted_to_bits)
    print("Cipher text : ", convertBinaryToHex(encrypted_to_bits))
```
